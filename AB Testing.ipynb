```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulate data for control group (no intervention)
control_group = np.random.normal(loc=30, scale=10, size=100)

# Simulate data for treatment group (with intervention)
treatment_group = np.random.normal(loc=35, scale=12, size=100)
```

```python
# Plot histograms for both groups
plt.hist(control_group, bins=20, alpha=0.5, label='Control Group')
plt.hist(treatment_group, bins=20, alpha=0.5, label='Treatment Group')
plt.xlabel('Metric Value')
plt.ylabel('Frequency')
plt.title('Distribution of Metric in Control and Treatment Groups')
plt.legend()
plt.show()
```

```python
# Perform t-test
t_stat, p_value = stats.ttest_ind(control_group, treatment_group)

# Set significance level
alpha = 0.05

# Print t-statistic and p-value
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# Compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between groups.")
```

*If the p-value is less than 0.05 (alpha level), we reject the null hypothesis, indicating a significant difference between the control and   treatment groups.
*Positive t-statistic suggests the treatment group's metric is higher, supporting the effectiveness of the intervention.
